```python
python_code_template = """
import duckdb
import time
import statistics
import pandas as pd

con = duckdb.connect()
con.execute("INSTALL tpcds")
con.execute("LOAD tpcds")

if {is_native}:
    scale_factor = 10
    con.execute(f"CALL dsdgen(sf={{scale_factor}})")
    print(f"TPC-DS data generated with scale factor {{scale_factor}}")
else:
    con.execute(f"ATTACH 'dbname=postgres user=postgres password= host=localhost' AS db (TYPE POSTGRES, SCHEMA 'public', READ_ONLY);")

    tables = [
        'call_center', 'catalog_page', 'catalog_returns', 'catalog_sales', 'customer',
        'customer_address', 'customer_demographics', 'date_dim', 'household_demographics',
        'income_band', 'inventory', 'item', 'promotion', 'reason', 'ship_mode', 'store',
        'store_returns', 'store_sales', 'time_dim', 'warehouse', 'web_page', 'web_returns',
        'web_sales', 'web_site'
    ]
    for table in tables:
        con.execute(f"CREATE VIEW {{table}} AS SELECT * FROM db.public.{{table}};")

queries = list(range(1, 100))

num_runs = 12

results = []

for query_num in queries:
    times = []
    for _ in range(num_runs):
        query_str = f"PRAGMA tpcds({{query_num}});";
        start_time = time.time()
        try:
            con.execute(query_str)
        except Exception as e:
            print(f"Query {{query_num}} failed: {{e}}")
            times.append(float('inf'))
        else:
            end_time = time.time()
            elapsed_time = end_time - start_time
            times.append(elapsed_time)

    avg_time = statistics.mean(times)
    results.append({{"Query Number": query_num, "Average Time (seconds)": avg_time}})
    print(f"Query {{query_num}} average time: {{avg_time:.4f}} seconds")

df = pd.DataFrame(results)

csv_file = "/tmp/tpcds_query_results.csv"
df.to_csv(csv_file, index=False)
print(f"Results saved to {{csv_file}}")
"""

with open('tpcds_duckdb_native.py', 'w') as file:
    file.write(python_code_template.format(is_native='True'))

with open('tpcds_duckdb_postgres.py', 'w') as file:
    file.write(python_code_template.format(is_native='False'))
```

```python
def draw_graph(filename, ylim=[0, 1000]):
    import pandas as pd
    import matplotlib.pyplot as plt

    df = pd.read_csv(filename)

    plt.figure(figsize=(15, 6))
    plt.bar(df['Query Number'], df['Average Time (seconds)'])
    plt.xlabel('Query Number')
    plt.ylabel('Average Time (seconds)')
    plt.title('TPC-DS Query Performance')
    plt.xticks(rotation=90)
    plt.ylim(ylim)
    plt.tight_layout()
    plt.show()
```

```python
import duckdb

con = duckdb.connect(database=':memory:')
query = "COPY (SELECT * FROM tpcds_queries()) TO 'tpcds_queries.json' (FORMAT JSON);"
con.execute(query)
```

```python
python_code_str = """
import psycopg2
import time
import pandas as pd
import json

def load_sql_queries_from_json(jsonl_file):
    queries = []
    with open(jsonl_file, mode='r', encoding='utf-8') as file:
        for line in file:
            query_data = json.loads(line)
            queries.append((query_data['query_nr'], query_data['query']))
    return queries

def measure_query_performance(conn, query, repetitions=3, timeout_ms=1800*1000):
    execution_times = []
    with conn.cursor() as cursor:
        cursor.execute(f"SET statement_timeout = {timeout_ms};")

        for _ in range(repetitions):
            start_time = time.time()
            try:
                cursor.execute(query)
            except:
                print(f"Query timed out after {timeout_ms/1000} seconds: {query}")
                conn.commit()
                return float('inf')

            end_time = time.time()
            execution_times.append(end_time - start_time)

    conn.commit()
    avg_time = sum(execution_times) / repetitions
    return avg_time

def main(json_file, output_csv):
    conn = psycopg2.connect('dbname=postgres user=postgres password= host=localhost')

    sql_queries = load_sql_queries_from_json(json_file)

    results = []

    for query_no, query in sql_queries:
        if int(query_no) <= 16:
            continue

        avg_time = measure_query_performance(conn, query)
        results.append({'query_no': query_no, 'avg_execution_time': avg_time})
        print(f"Query {query_no} average time: {avg_time:.4f} seconds")

    df = pd.DataFrame(results)
    df.to_csv(output_csv, index=False)

    conn.close()

if __name__ == "__main__":
    main('tpcds_queries.json', '/tmp/tpcds_query_results.csv')
"""

with open('tpcds_postgres.py', 'w') as file:
    file.write(python_code_str)
```

```python
def draw_graph_comparison(filenames, labels):
    import pandas as pd
    import matplotlib.pyplot as plt
    plt.figure(figsize=(15, 6))

    width = 0.25
    x = range(len(pd.read_csv(filenames[0])['Query Number']))
    for i, filename in enumerate(filenames):
        df = pd.read_csv(filename)
        plt.bar([xi + i * width for xi in x], df['Average Time (seconds)'], width=width, label=labels[i])

    plt.xlabel('Query Number')
    plt.ylabel('Average Time (seconds)')
    plt.title('TPC-DS Query Performance Comparison')
    plt.xticks(rotation=90)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
filenames = ['tpcds_query_results_duckdb_native.csv', 'tpcds_query_results_postgres.csv']
labels = ['DuckDB', 'Postgres']
draw_graph_comparison(filenames, labels)
```

```python
filenames = ['tpcds_query_results_duckdb_postgres.csv', 'tpcds_query_results_postgres.csv']
labels = ['DuckDB w/Postgres', 'Postgres']
draw_graph_comparison(filenames, labels)
```
